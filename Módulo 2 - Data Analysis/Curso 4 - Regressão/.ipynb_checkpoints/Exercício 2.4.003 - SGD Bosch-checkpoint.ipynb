```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from IPython.display import Image, display
```

```python
caminho = "dataset/"
arquivo = "ToolDataSheetVsNQuebras.csv" #<- arquivo a ser utilizado
```

### Variável dependente (para predição -y-):
*Número de quebras por ano*: `nQuebras`

### Chave primária: 
*código do tipo da ferramenta*:  `PartNumber`

### Variáveis independentes:
*geometria e parâmetros da ferramenta de corte* 

1. `LargerDiameter`
2. `StemDiameter`
3. `TotalLength`
4. `ChannelLength`

```python
# display(Image(caminho+"usinagem_geo.PNG"))
```

```python
dadosB = None

try:  # carrega dos dados , o separador é ";" pois tem campos na coluna Supplier com ","
    dadosB = pd.read_csv(caminho+arquivo, sep=";")
except IOError as err:
    print (err)
    
dadosB = dadosB[dadosB.PropellerAngle <= 360] #remove os demais dados incorretos
print(dadosB.shape)
dadosB.sample(5).T
```

```python
# vamos substituir os valores vazios por uma anotação e criar uma nova coluna para marcar um flag específico para isso
dadosB.MaterialClass.fillna('NA', inplace=True)
dadosB['MissMaterialClass'] = 0
dadosB.loc[dadosB.MaterialClass == 'NA', 'MissMaterialClass'] = 1

print(dadosB.Family.unique())
print(dadosB.Supplier.unique())
print(dadosB.MaterialClass.unique())
```

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()

for col in ['Family', 'Supplier', 'MaterialClass']:
    dadosB[col] = le.fit_transform(dadosB[col])
    
dadosB.sample(5)
```

```python
# criei essa função para automatizar a preparação das colunas de um df para regressão

import analise_exploratoria as ae
import importlib
importlib.reload(ae)
```

```python
index_cols = ['PartNumber']
y_cols = ['nQuebras']
ig_cols = ['PartNumber']

X, y, M = ae.analise_exploratoria(dadosB, y_cols, ig_cols, graph=True)
```

```python
# vemos alguns outliers bem específicos acima, vamos testar sem estes valores (vemos apenas 1 observação outlier para cada)
dadosB = dadosB[dadosB.LargerDiameter < 100]
dadosB = dadosB[dadosB.PropellerAngle < 200]
dadosB = dadosB[dadosB.MissMaterialClass != 1]
```

```python
# o flag criado para MissMaterialClass não foi relevante
# também vimos uma alta correlação de StemDiameter e ChannelLength com outras variáveis
# removendo as 2 também da modelagem para evitar problemas de colinearidade

ig_cols = ['PartNumber', 'MissMaterialClass', 'StemDiameter', 'ChannelLength']
X, y, M = ae.analise_exploratoria(dadosB, y_cols, ig_cols, graph=True)
```

```python
# modelando 100% dos dados, sem split de dataset

from sklearn.linear_model import SGDRegressor
from sklearn.preprocessing import MinMaxScaler
from sklearn.preprocessing import StandardScaler

#scaler = MinMaxScaler()
scaler = StandardScaler()
scaler.fit(X)
X_scaled = scaler.transform(X)

reg_grad = SGDRegressor(max_iter=10000, warm_start=True)
reg_grad.fit(X_scaled, y.values.ravel())
y_hat = reg_grad.predict(X_scaled)

plt.scatter(y, y_hat, color='g')
plt.xlabel('Observado')
plt.ylabel('Predito')
```

```python
# os valores de X foram normalizados, vemos que estão ok após a transformação
import seaborn as sns
for i in range(X_scaled.shape[1]):
    sns.kdeplot(X_scaled[:][i])
plt.show()
```

```python
# criei essa função para automatizar a avaliação do resultado de uma modelagem multivariada, usando critérios estatísticos

import analise_residuos as ar
import importlib
importlib.reload(ar)
```

```python
ar.analise_residuos(y, y_hat, graph=True)
```

```python
# vamos fazer algumas repetições no método do SGD e avaliar se há alguma melhoria

fig, ax = plt.subplots(3, 3, figsize=(16,12))

for i in range(9):
    j, k = i//3, i%3
    reg_grad.fit(X_scaled, y.values.ravel())
    y_hat = reg_grad.predict(X_scaled)
    ax[j][k].scatter(y, y_hat, color='g')
    ax[j][k].set_xlabel('Observado')
    ax[j][k].set_ylabel('Predito')
    ar.analise_residuos(y, y_hat, False)
plt.show()
```

```python
# os modelos não estão satisfatórios, os resíduos ferem diversas premissas e os valores de R2 e RSS bastante precários

# vamos comparar com um ajuste linear
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import MinMaxScaler
from sklearn.preprocessing import StandardScaler

reg_linear = LinearRegression()
reg_linear.fit(X_scaled, y.values.ravel())
y_hat = reg_linear.predict(X_scaled)

plt.scatter(y, y_hat, color='g')
plt.xlabel('Observado')
plt.ylabel('Predito')
```

```python
# idem, resultado precário também para o modelo linear
```

```python
# vamos testar mesmo assim como ficaria um treinamento usando train/test

from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20)

#scaler = MinMaxScaler()
scaler = StandardScaler()

# vamos calcular os parâmetros mi e sigma para o X de treino e armazenar no objeto scaler
scaler.fit(X_train)

# vamos aplicar o modelo de transformação para os conjuntos de treino e teste
X_train_scaled = scaler.transform(X_train)
X_test_scaled = scaler.transform(X_test)

# aplicando o SGD
reg_grad = SGDRegressor(max_iter=10000, warm_start=True)
reg_grad.fit(X_train_scaled, y_train.values.ravel())
y_hat = reg_grad.predict(X_test_scaled)

plt.scatter(y_test, y_hat, color='g')
plt.xlabel('Observado')
plt.ylabel('Predito')
```

```python
ar.analise_residuos(y_test, y_hat, graph=True)
```

```python
# novamente, e como já era esperado (mais graus de liberdade ao separar os 2 subconjuntos), o desempenho piorou

# R2 negativo, resíduo não normal, incapacidade de prever valores acima de 100 (domínio vai até 600)
# o histograma acima dá uma dica de que talvez uma transformação não linear em y possa fazer sentido, vamos testar
```

```python
# vamos usar a transformação de boxcox para transformar nosso y, armazenando o valor de lambda na variável lmbda
# isso é importante para reverter os resultados à dimensão de origem

from scipy import stats

yt_train, lmbda = stats.boxcox(y_train)
print('Lambda = ',lmbda)
yt_test = stats.boxcox(y_test, lmbda=lmbda)

plt.hist(yt_train)
plt.hist(yt_test)
plt.show()
```

```python
# aplicando o SGD
reg_grad = SGDRegressor(max_iter=10000, warm_start=True)
reg_grad.fit(X_train_scaled, yt_train.ravel())
yt_hat = reg_grad.predict(X_test_scaled)

plt.scatter(yt_test, yt_hat, color='g')
plt.xlabel('Observado')
plt.ylabel('Predito')
```

```python
ar.analise_residuos(yt_test.ravel(), yt_hat, graph=True)
```

```python
# agora observamos uma grande melhoria nos resultados, com redução pela metade do valor do R2
# também a análise dos resíduos atende satisfatoriamente a todas as premissas:
#  - resíduo estocástico, homocedástico e normalmente distribuído

# entretanto o valor de R2 continua bastante insatisfatório
# entendo que alguma informação relevante não está sendo considerada neste dataset, falta alguma informação
# precisaria entender melhor o processo para propor outras melhorias e modelagens
```

```python
# vamos fazer algumas repetições no método do SGD e avaliar se há alguma melhoria

fig, ax = plt.subplots(3, 3, figsize=(16,12))

for i in range(9):
    j, k = i//3, i%3
    reg_grad.fit(X_train_scaled, yt_train.ravel())
    yt_hat = reg_grad.predict(X_test_scaled)
    ax[j][k].scatter(yt_test.ravel(), yt_hat, color='g')
    ax[j][k].set_xlabel('Observado')
    ax[j][k].set_ylabel('Predito')
    ar.analise_residuos(yt_test.ravel(), yt_hat, False)
plt.show()
```
