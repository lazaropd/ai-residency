*  uma image de 427x640 pixels, sendo cada pixel trs valores correspondente ao RGB

* Os valores de RGB variam de 0 a 255. Dessa forma, os valores nessa matriz so entre 0 a 255, dividindo todo valor por 255, obtemos uma escala entre 0 a 1
* Em seguida os dados foram redimensionados, de 2D para 1D (apenas mudana de posio dos dados, no altera o contedo)

```python
data = china / 255.0 # Refazendo para escala entre 0 a 1
data = data.reshape(427 * 640, 3) # transformando o formato de dados 2D para 1D, apenas mudana de representao
data.shape
```

```python
print(data)
```

### Funo para plotar o valor de cada pixel em duas dimenses

```python
import numpy as np

def plot_pixels(data, title, colors=None, N=10000):
    if colors is None:
        colors = data
    
    # choose a random subset
    rng = np.random.RandomState(0)
    print(rng)
    i = rng.permutation(data.shape[0])[:N]
    print(i)
    colors = colors[i]
    R, G, B = data[i].T
    
    fig, ax = plt.subplots(1, 2, figsize=(16, 6))
    ax[0].scatter(R, G, color=colors, marker='.')
    ax[0].set(xlabel='Red', ylabel='Green', xlim=(0, 1), ylim=(0, 1))

    ax[1].scatter(R, B, color=colors, marker='.')
    ax[1].set(xlabel='Red', ylabel='Blue', xlim=(0, 1), ylim=(0, 1))

    fig.suptitle(title, size=20);
```

```python
plot_pixels(data, title='Input color space: 16 million possible colors')
```

### Usando Kmeans para encontrar 16 grupos
* O predict est sendo usado para colocar todos os itens em um dos 16 clusters

```python
data
```

```python
from sklearn.cluster import MiniBatchKMeans
kmeans = MiniBatchKMeans(16)
kmeans.fit(data)

kmeans.cluster_centers_
```

```python
new_colors = kmeans.cluster_centers_[kmeans.predict(data)]

plot_pixels(new_colors, colors=new_colors, title="Reduced color space: 16 colors")
```

```python
new_colors
```

```python
china_recolored = new_colors.reshape(china.shape)

fig, ax = plt.subplots(1, 2, figsize=(16, 6),
                       subplot_kw=dict(xticks=[], yticks=[]))
fig.subplots_adjust(wspace=0.05)
ax[0].imshow(china)
ax[0].set_title('Original Image', size=16)
ax[1].imshow(china_recolored)
ax[1].set_title('16-color Image', size=16);
```

### PCA Principal Component Analysis
* Varincia e covarincia so uma medida do espalhamento de um conjunto de pontos ao redor do centro de massa (mdia)
* Varincia - medida do desvio da mdia para pontos em uma dimenso, por exemplo alturas
* Covarincia como uma medida de quanto cada uma das dimenses variam da mdia em relao um ao outro.
* A covarincia  medida entre 2 dimenses para verificar se existe uma relao entre as 2 dimenses. 
    * Ex:  nmero de horas estudado e notas obtidas.
* A covarincia entre uma dimenso e ela mesma  a variancia


### Matriz de covarincia
*  uma matriz quadrada de tamanho igual a quantidade de linhas
* A diagonal principal mostra a varincia


### Interpretando covarincia

* Um valor positivo de covarincia indica ambos dimenses aumentam ou diminuem juntas. 

* Um valor negativo indica enquanto um aumenta a outras diminuies ou vice-versa 

* Se a covarincia for zero: as duas dimenses so independentes um do outro

* Clculos de covarincia so usados para encontrar relaes em conjuntos de dados com muitas  dimenses (geralmente maiores que 3) onde a visualizao  difcil

### O exemplo de dados abaixo mostra um dataset com trs colunas
* As linhas so uma funo da primeira linhas multiplicada por um escalar
* A Seguir a matriz de coraviancia

```python
x = np.array([[1, 2,3], [2, 4,6], [4, 8, 12], [3, 6, 9], [3, 6, 9], [5, 10, 15], [6, 12, 18]])
print(x)
print(np.cov(x)) # covariao
```

### A figura abaixo mostra uma linha com os pontos determinados pelas 3 dimenses

* Considere um novo sistema de coordenadas em que um dos eixos est na direo dessa linha
* Um dos eixos pode entao ser usado para definir os outros dois

* Anlise de componentes principais (PCA)  uma tcnica usada para simplificar um conjunto de dados
*  uma transformao linear que escolhe uma nova sistema de coordenadas para o conjunto de dados maior variao por qualquer projeo dos dados 
* O conjunto com maior variancia fica no primeiro eixo (ento chamado de primeiro componente principal),
    * A segunda maior varincia no segundo eixo, e assim por diante.
* O PCA pode ser usado para reduzir a dimensionalidade, eliminando os componentes principais posteriores.

#### Considere o conjunto de pontos a seguir
* 200 pontos aleatrios em duas dimenses

```python
rng = np.random.RandomState(1)
X = np.dot(rng.rand(2, 2), rng.randn(2, 200)).T
plt.scatter(X[:, 0], X[:, 1])
plt.axis('equal');
```

* H uma relao  linear entre as variveis x1 e x2 (ou x e y por exemplo) 


* O problema de aprendizado no supervisionado tenta aprender sobre a relao entre os valores x e y. 
    * Contrapondo a tentar prever os valores de y a partir dos valores de x

* Na anlise de componentes principais (PCA), esse relacionamento  quantificado encontrando uma lista dos eixos principais nos dados e usando esses eixos para descrever o conjunto de dados. 

* Usando o estimador PCA do Scikit-Learn:

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=2)
pca.fit(X)
```

```python
print(pca.components_)
```

```python
print(pca.explained_variance_)
```

```python
def draw_vector(v0, v1, ax=None):
    ax = ax or plt.gca()
    arrowprops=dict(arrowstyle='->',
                    linewidth=2,
                    shrinkA=0, shrinkB=0)
    ax.annotate('', v1, v0, arrowprops=arrowprops)

# plot data
plt.scatter(X[:, 0], X[:, 1], alpha=0.2)
for length, vector in zip(pca.explained_variance_, pca.components_):
    v = vector * 3 * np.sqrt(length)
    draw_vector(pca.mean_, pca.mean_ + v)
plt.axis('equal');
```

Esses vetores representam os principais eixos dos dados, e o comprimento do vetor  uma indicao de quo "importante" esse eixo  na descrio da distribuio dos dados 
* Mais precisamente,  uma medida da variao dos dados quando projetados. 
* Nesse eixo. A projeo de cada ponto de dados nos eixos principais so os "componentes principais" dos dados.

* O uso do PCA para reduo de dimensionalidade envolve zerar um ou mais dos menores componentes principais
 * Resultando em uma projeo em dimenso mais baixa dos dados que preserva a variao mxima dos dados.
* Abaixo  mostrado o uso do PCA

```python
#pca = PCA(n_components=1)
pca = PCA(0.1)

pca.fit(X)
X_pca = pca.transform(X)
print("original shape:   ", X.shape)
print("transformed shape:", X_pca.shape)
```

* Os dados transformados foram reduzidos para uma nica dimenso. 
* Podemos realizar a transformao inversa desses dados reduzidos
* Grfico comparando dados transformados com dados reduzidos

```python
X_new = pca.inverse_transform(X_pca)
plt.scatter(X[:, 0], X[:, 1], alpha=0.2)
plt.scatter(X_new[:, 0], X_new[:, 1], alpha=0.5)
plt.axis('equal');
```

```python
print(pca.explained_variance_.sum())
```

Este conjunto de dados de dimenso reduzida , em alguns sentidos, "bom o suficiente" para codificar os relacionamentos mais importantes entre os pontos: apesar de reduzir a dimenso dos dados em 50%, o relacionamento geral entre os pontos de dados  principalmente preservado.

### Exemplo de uso de PCA para identificao de faces

```python
from sklearn.datasets import fetch_lfw_people
faces = fetch_lfw_people(min_faces_per_person=60)
print(faces.target_names)
print(faces.target.shape)
print(faces.images.shape)
print(faces.data.shape)
```

```python
pca = PCA(150)
pca.fit(faces.data)
```

```python
fig, axes = plt.subplots(3, 8, figsize=(9, 4),
                         subplot_kw={'xticks':[], 'yticks':[]},
                         gridspec_kw=dict(hspace=0.1, wspace=0.1))
for i, ax in enumerate(axes.flat):
    ax.imshow(pca.components_[i].reshape(62, 47), cmap='bone')
plt.show()
```

```python
plt.plot(np.cumsum(pca.explained_variance_ratio_))
plt.xlabel('number of components')
plt.ylabel('cumulative explained variance');
```

```python
# Compute the components and projected faces
components = pca.transform(faces.data)
projected = pca.inverse_transform(components)
```

```python
# Plot the results
fig, ax = plt.subplots(2, 10, figsize=(10, 2.5),
                       subplot_kw={'xticks':[], 'yticks':[]},
                       gridspec_kw=dict(hspace=0.1, wspace=0.1))
for i in range(10):
    ax[0, i].imshow(faces.data[i].reshape(62, 47), cmap='binary_r')
    ax[1, i].imshow(projected[i].reshape(62, 47), cmap='binary_r')
    
ax[0, 0].set_ylabel('full-dim\ninput')
ax[1, 0].set_ylabel('150-dim\nreconstruction');
```

### Testes usando PCA antes de classificar

```python
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import confusion_matrix, classification_report

data = load_breast_cancer()
print(data.target.shape)
print(data.data.shape)
X=data.data
y=data.target

X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.1)

sgd_clf = KNeighborsClassifier(n_neighbors=3)

sgd_clf.fit(X_train, y_train)

y_pred = sgd_clf.predict(X_test)

cm = confusion_matrix(y_test, y_pred)

print(cm)
print(classification_report(y_test, y_pred))
```

```python
pca = PCA(2)
pca.fit(X)
XPCA = pca.transform(X)

print(XPCA.shape)
fig, ax = plt.subplots(1, 2, figsize=(16, 6))
ax[0].scatter(XPCA[:,0], XPCA[:,1])

y=data.target

X_train, X_test, y_train, y_test = train_test_split(XPCA,y, test_size=0.1)

sgd_clf = KNeighborsClassifier(n_neighbors=3)

sgd_clf.fit(X_train, y_train)

y_pred = sgd_clf.predict(X_test)

cm = confusion_matrix(y_test, y_pred)

print(cm)
print(classification_report(y_test, y_pred))
```

### Pipeline de classificao com PCA

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.tree import DecisionTreeClassifier 
from sklearn.model_selection import GridSearchCV

#pipe_knn = Pipeline([('scl', StandardScaler()), ('clf', KNeighborsClassifier())])
pipetree = Pipeline([ ('scl', StandardScaler()), ('pca', PCA(0.8)), ('clf', DecisionTreeClassifier())])

#pipe = [pipe_knn, pipetre]
pipe = [pipetree]

param_range = [1, 2,3 ,4]

# grid search params
#grid_params = [{'clf__criterion': ['gini', 'entropy'],
#               'clf__presort': [True, False]}]
grid_params = [{'clf__criterion': ['gini', 'entropy'],
    'clf__min_samples_leaf': param_range,
    'clf__max_depth': param_range,
    'clf__min_samples_split': param_range[1:]}]

# Construct grid search
gs = GridSearchCV(estimator=pipetree,
    param_grid=grid_params,
    scoring='accuracy')

# Fit using grid search
gs.fit(X_train, y_train)

# Best accuracy
print('Best accuracy: %.3f' % gs.best_score_)

# Best params
print('\nBest params:\n', gs.best_params_)
```
