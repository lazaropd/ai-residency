```python
import numpy as np
import matplotlib.pyplot as plt 

from sklearn.datasets import fetch_lfw_people
from sklearn.decomposition import PCA as RandomizedPCA

from sklearn.model_selection import train_test_split

from sklearn import svm
from sklearn import metrics

from tensorflow.keras.layers import Input, Dense 
from tensorflow.keras.models import Sequential, Model 
```

## Método Tradicional: PCA

```python
NCOMPONENTS = 150

faces = fetch_lfw_people(min_faces_per_person=60)
print('Classes:', faces.target_names)
print('Dataset shape:', faces.images.shape)

X_train, X_test, y_train, y_test = train_test_split(faces.data,
        faces.target, random_state=0)
print('Split de treino e teste:', X_train.shape, X_test.shape)

pca = RandomizedPCA(n_components=NCOMPONENTS, svd_solver='randomized', whiten=True)
pca.fit(faces.data)
```

```python
fig, axes = plt.subplots(3, 8, figsize=(9, 4),
                         subplot_kw={'xticks':[], 'yticks':[]},
                         gridspec_kw=dict(hspace=0.1, wspace=0.1))
for i, ax in enumerate(axes.flat):
    ax.imshow(pca.components_[i].reshape(62, 47), cmap='bone')  
plt.suptitle('Amostra de 24 das PCs extraídas')
plt.show()
    
plt.plot(np.cumsum(pca.explained_variance_ratio_))
plt.plot([0, NCOMPONENTS], [1., 1.], color='red', lw=1)
plt.xlabel('Número de Componentes')
plt.ylabel('Variância Acumulada')
plt.title('Explicação Cumulativa da Variância')
plt.show()
```

```python
pca = RandomizedPCA(NCOMPONENTS).fit(faces.data)
components = pca.transform(faces.data)
projected = pca.inverse_transform(components)
print('Shape do espaço original:', faces.data.shape)
print('Shape do espaço reduzido:', components.shape)
print('Shape do espaço reconstruído:', projected.shape)
```

```python
fig, ax = plt.subplots(2, 10, figsize=(10, 2.5),
                       subplot_kw={'xticks':[], 'yticks':[]},
                       gridspec_kw=dict(hspace=0.1, wspace=0.1))
for i in range(10):
    ax[0, i].imshow(faces.data[i].reshape(62, 47), cmap='binary_r')
    ax[1, i].imshow(projected[i].reshape(62, 47), cmap='binary_r')
    
ax[0, 0].set_ylabel('full-dim\ninput')
ax[1, 0].set_ylabel('N-dim\nreconstruction');
```

## Comparação de desempenho usando um classificador SVM

```python
# treino sem o uso da PCA -> ndim >> sample

print('Espaço de treino:', X_train.shape, X_test.shape)
clf = svm.SVC(C=5., gamma=0.001)
clf.fit(X_train, y_train)
y_pred = clf.predict(X_test)
print(metrics.classification_report(y_test, y_pred))
```

```python
# treino usando a PCA -> ndim < size

pca = RandomizedPCA(n_components=NCOMPONENTS, whiten=True)
pca.fit(X_train)
X_train_pca = pca.transform(X_train)
X_test_pca = pca.transform(X_test)

print('Espaço de treino:', X_train_pca.shape, X_test_pca.shape)
clf = svm.SVC(C=5., gamma=0.001)
clf.fit(X_train_pca, y_train)
y_pred = clf.predict(X_test_pca)
print(metrics.classification_report(y_test, y_pred))
```

## Método Alternativo: Autoencoder

```python
latent_dim = NCOMPONENTS # dimensão do espaço latente

input_shape = 2914

# construção do codificador
encoder = Sequential(name="encoder") 
encoder.add(Dense(128, input_shape=(input_shape,), activation="relu")) 
#encoder.add(Dense(256, activation="relu"))
#encoder.add(Dense(512, activation="relu"))
encoder.add(Dense(latent_dim)) # Vetor Latente
#encoder.summary()

# construção do decodificador
decoder = Sequential(name="decoder")
decoder.add(Dense(128, input_shape=(latent_dim,), activation="relu"))
#decoder.add(Dense(256, activation="relu"))
#decoder.add(Dense(512, activation="relu"))
decoder.add(Dense(input_shape, activation="sigmoid")) 
#decoder.summary()

# construção do modelo consolidado
input_img = Input(shape=(input_shape,) ) 
z = encoder(input_img) # obtem vetor latente através da entrada
recons = decoder(z) # reconstrói a imagem a partir do vetor latente
ae = Model(input_img, recons) # cria o modelo combinando encoder e decoder
#ae.summary()

ae.compile(optimizer='Adam', loss='categorical_crossentropy', metrics=["categorical_crossentropy"])

X_train_scaled = (X_train / 255.).astype(np.float32)
X_test_scaled  = ( X_test / 255.).astype(np.float32)

batch_size = 16
epochs = 100
history = ae.fit(X_train_scaled, X_train_scaled, batch_size=batch_size, epochs=epochs, 
                 validation_data=(X_test_scaled, X_test_scaled), verbose=0)
print('Loss média nas 5 últimas épocas:', np.mean(history.history['val_loss'][-5:]))

fig, ax = plt.subplots(2, 10, figsize=(10, 2.5),
                       subplot_kw={'xticks':[], 'yticks':[]},
                       gridspec_kw=dict(hspace=0.1, wspace=0.1))
for i in range(10):
    ax[0, i].imshow(X_train_scaled[i].reshape(62, 47), cmap='binary_r')
    imgb = np.reshape(X_train_scaled[i], (1, input_shape)) 
    recons_img = ae.predict(imgb).reshape(62, 47)
    ax[1, i].imshow(recons_img, cmap='binary_r')
    
ax[0, 0].set_ylabel('full-dim\ninput')
ax[1, 0].set_ylabel('N-dim\nreconstruction')
plt.suptitle('Reconstrução das imagens de treino')
plt.show()

fig, ax = plt.subplots(2, 10, figsize=(10, 2.5),
                       subplot_kw={'xticks':[], 'yticks':[]},
                       gridspec_kw=dict(hspace=0.1, wspace=0.1))
for i in range(10):
    ax[0, i].imshow(X_test_scaled[i].reshape(62, 47), cmap='binary_r')
    imgb = np.reshape(X_test_scaled[i], (1, input_shape)) 
    recons_img = ae.predict(imgb).reshape(62, 47)
    ax[1, i].imshow(recons_img, cmap='binary_r')
    
ax[0, 0].set_ylabel('full-dim\ninput')
ax[1, 0].set_ylabel('N-dim\nreconstruction')
plt.suptitle('Reconstrução das imagens de teste')
plt.show()
```

```python
# criando um modelo bem simples, vamos comparar o desempenho de X e reduced

NCLASSES = len(faces.target_names)

model = Sequential() 
model.add(Dense(256, input_shape=(input_shape,), activation="relu")) 
model.add(Dense(NCLASSES, activation='softmax')) 
model.summary()
```

```python
epochs = 100
model.compile(optimizer='Adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])

history = model.fit(X_train_scaled, y_train, batch_size=batch_size, epochs=epochs, 
                    validation_data=(X_test_scaled, y_test), verbose=0)

plt.plot(history.history['accuracy'], color='blue', label='Treino')
plt.plot(history.history['val_accuracy'], color='orange', label='Teste')
plt.ylim(0., 1.)
plt.legend()
plt.title('Evolução da acurácia - Espaço original')
plt.show()

y_pred = np.argmax(model.predict(X_test_scaled), axis=1)
print(metrics.classification_report(y_test, y_pred))
```

```python
X_train_reduced = ae.predict(X_train_scaled)
X_test_reduced = ae.predict(X_test_scaled)
```

```python
epochs = 100
model.compile(optimizer='Adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])

history = model.fit(X_train_reduced, y_train, batch_size=batch_size, epochs=epochs, 
                    validation_data=(X_test_reduced, y_test), verbose=0)

plt.plot(history.history['accuracy'], color='blue', label='Treino')
plt.plot(history.history['val_accuracy'], color='orange', label='Teste')
plt.ylim(0., 1.)
plt.legend()
plt.title('Evolução da acurácia - Espaço reconstruído')
plt.show()

y_pred = np.argmax(model.predict(X_test_reduced), axis=1)
print(metrics.classification_report(y_test, y_pred))
```
