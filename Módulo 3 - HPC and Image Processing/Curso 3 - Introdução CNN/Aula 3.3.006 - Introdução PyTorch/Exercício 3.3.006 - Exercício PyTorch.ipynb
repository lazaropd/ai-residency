## 1. Fashion MNIST

```python
import pandas as pd
import numpy as np
from matplotlib import pyplot

from sklearn import preprocessing
from sklearn.preprocessing import LabelEncoder, MinMaxScaler
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split

from torch.utils.data import Dataset, DataLoader, random_split

from torch import Tensor

from torch.nn import Linear, Conv2d, MaxPool2d
from torch.nn import ReLU
from torch.nn import Sigmoid, Softmax
from torch.nn import Module

from torch.nn import BCELoss, CrossEntropyLoss, MSELoss
from torch.nn.init import kaiming_uniform_, xavier_uniform_
from torch.optim import SGD

from torchvision.transforms import Resize, Compose, ToTensor, Normalize
from torchvision.datasets import ImageFolder

import utils
```

```python
from torchvision.datasets import FashionMNIST

# define location to save or load the dataset
path = '~/.torch/datasets/fashionmnist'
batch_size = 32

# define the transforms to apply to the data
trans = Compose([ToTensor()])

# download and define the datasets
train = FashionMNIST(path, train=True, download=True, transform=trans)
test = FashionMNIST(path, train=False, download=True, transform=trans)

# define how to enumerate the datasets
train_dl = DataLoader(train, batch_size=batch_size, shuffle=True)
test_dl = DataLoader(test, batch_size=batch_size, shuffle=True)

# get one batch of images
i, (inputs, targets) = next(enumerate(train_dl))
print(i)

# plot some images
fig, axes = pyplot.subplots(figsize=(15,8))
for i in range(25):
    pyplot.subplot(5, 5, i+1)
    pyplot.imshow(inputs[i][0], cmap='gray')
pyplot.show()
```

```python
class CNN(Module):
    def __init__(self, n_channels):
        super(CNN, self).__init__()
        # input to first hidden layer
        self.hidden1 = Conv2d(n_channels, 32, (3,3))
        kaiming_uniform_(self.hidden1.weight, nonlinearity='relu')
        self.act1 = ReLU()
        self.hidden2 = Conv2d(32, 32, (3,3))
        kaiming_uniform_(self.hidden2.weight, nonlinearity='relu')
        self.act2 = ReLU()
        self.pool1 = MaxPool2d((2,2), stride=(2,2))
        # second hidden layer
        self.hidden3 = Conv2d(32, 64, (3,3))
        kaiming_uniform_(self.hidden3.weight, nonlinearity='relu')
        self.act3 = ReLU()
        self.pool2 = MaxPool2d((2,2), stride=(2,2))
        # fully connected layer
        self.hidden4 = Linear(5*5*64, 400)
        kaiming_uniform_(self.hidden4.weight, nonlinearity='relu')
        self.act4 = ReLU()
        self.hidden5 = Linear(400, 100)
        kaiming_uniform_(self.hidden5.weight, nonlinearity='relu')
        self.act5 = ReLU()
        # output layer
        self.hidden6 = Linear(100, 10)
        xavier_uniform_(self.hidden6.weight)
        self.act6 = Softmax(dim=1)
 
    def forward(self, X):
        X = self.hidden1(X)
        X = self.act1(X)
        X = self.hidden2(X)
        X = self.act2(X)
        X = self.pool1(X)
        X = self.hidden3(X)
        X = self.act3(X)
        X = self.pool2(X)
        # flatten
        X = X.view(-1, 4*8*50)
        X = self.hidden4(X)
        X = self.act4(X)
        X = self.hidden5(X)
        X = self.act5(X)
        X = self.hidden6(X)
        X = self.act6(X)
        return X 

# train the model
def train_model(train_dl, model, epochs):
    criterion = CrossEntropyLoss()
    optimizer = SGD(model.parameters(), lr=0.01, momentum=0.9)
    for epoch in range(epochs):
        for i, (predictors, y_true) in enumerate(train_dl):
            optimizer.zero_grad()
            y_hat = model(predictors)
            loss = criterion(y_hat, y_true)
            loss.backward()
            optimizer.step()
        if epoch % (epochs/10) == 0: print('Current loss of %.2f after %d epochs' % (loss, epoch+1))

def evaluate_model(test_dl, model):
    predictions, actuals = list(), list()
    for i, (predictors, y_true) in enumerate(test_dl):
        y_true = y_true.numpy()
        y_true = y_true.reshape((len(y_true), 1))
        y_hat = model(predictors)
        y_hat = y_hat.detach().numpy()
        y_hat = np.argmax(y_hat, axis=1)
        y_hat = y_hat.reshape((len(y_hat), 1))
        y_hat = y_hat.round()
        predictions.append(y_hat)
        actuals.append(y_true)        
    predictions, actuals = np.vstack(predictions), np.vstack(actuals)
    acc = accuracy_score(actuals, predictions)
    return acc

def prepare_fashionMNIST_data(path, trans, batch_size):
    train = FashionMNIST(path, train=True, download=True, transform=trans)
    test = FashionMNIST(path, train=False, download=True, transform=trans)
    train_dl = DataLoader(train, batch_size=batch_size, shuffle=True)
    test_dl = DataLoader(train, batch_size=batch_size, shuffle=False)    
    return train_dl, test_dl



path = '~/.torch/datasets/fashionmnist'
trans = Compose([ToTensor(), Normalize((0.1307,), (0.3081,))])
batch_size = 32
epochs = 5

train_dl, test_dl = prepare_fashionMNIST_data(path, trans, batch_size)

model = CNN(1)
train_model(train_dl, model, epochs)
acc = evaluate_model(test_dl, model)
print('Accuracy: %.3f' % acc)
```

## Case Transport

```python
class MLP(Module):
    def __init__(self, n_inputs):
        super(MLP, self).__init__()
        self.hidden1 = Linear(n_inputs, 512)
        kaiming_uniform_(self.hidden1.weight, nonlinearity='relu')
        self.act1 = ReLU()
        self.hidden2 = Linear(512, 256)
        kaiming_uniform_(self.hidden2.weight, nonlinearity='relu')
        self.act2 = ReLU()
        self.hidden3 = Linear(256, 64)
        kaiming_uniform_(self.hidden3.weight, nonlinearity='relu')
        self.act3 = ReLU()
        self.hidden4 = Linear(64, 16)
        kaiming_uniform_(self.hidden4.weight, nonlinearity='relu')
        self.act4 = ReLU()
        self.hidden5 = Linear(16, 4)
        xavier_uniform_(self.hidden5.weight)
        self.act5 = Softmax(dim=1)
 
    # propagação da entrada pelas camadas
    def forward(self, X):
        X = self.hidden1(X)
        X = self.act1(X)
        X = self.hidden2(X)
        X = self.act2(X)
        X = self.hidden3(X)
        X = self.act3(X)
        X = self.hidden4(X)
        X = self.act4(X)
        X = self.hidden5(X)
        X = self.act5(X)
        return X

def train_model(train_dl, model, epochs):
    # define loss
    criterion = CrossEntropyLoss()
    optimizer = SGD(model.parameters(), lr=0.01, momentum=0.9)
    for epoch in range(epochs):
        for i, (predictors, y_true) in enumerate(train_dl):
            optimizer.zero_grad()
            y_hat = model(predictors)
            loss = criterion(y_hat, y_true.squeeze_())
            loss.backward()
            optimizer.step()
        if epoch % (epochs/10) == 0: print('Current loss of %.2f after %d epochs' % (loss, epoch+1))
    
def evaluate_model(test_dl, model):
    predictions, actuals = list(), list()
    for i, (predictors, y_true) in enumerate(test_dl):
        y_true = y_true.numpy()
        y_true = y_true.reshape((len(y_true), 1))
        y_hat = model(predictors)
        y_hat = y_hat.detach().numpy()
        y_hat = np.argmax(y_hat, axis=1)
        y_hat = y_hat.reshape((len(y_hat), 1))
        y_hat = y_hat.round()
        predictions.append(y_hat)
        actuals.append(y_true)        
    predictions, actuals = np.vstack(predictions), np.vstack(actuals)
    acc = accuracy_score(actuals, predictions)
    return acc    
    
class CSVDataset(Dataset):
    def __init__(self, path):
        df = pd.read_csv(path)
        self.X = df.values[:, :-1]
        self.y = df.values[:, -1]
        # ensure input data is floats
        self.X = self.X.astype('float32')
        self.y = LabelEncoder().fit_transform(self.y)
        self.y = self.y
        self.y = self.y.reshape((len(self.y), 1))
 
    def __len__(self):
        return len(self.X)
 
    def __getitem__(self, idx):
        return [self.X[idx], self.y[idx]]
 
    def get_splits(self, n_test=0.20):
        test_size = round(n_test * len(self.X))
        train_size = len(self.X) - test_size
        return random_split(self, [train_size, test_size])    

def prepare_data(path, batch_size):
    dataset = CSVDataset(path)
    train, test = dataset.get_splits()
    train_dl = DataLoader(train, batch_size=batch_size, shuffle=True)
    test_dl = DataLoader(test, batch_size=batch_size, shuffle=False)   
    print('Dataset length (total, train, test):', len(dataset), len(train_dl.dataset), len(test_dl.dataset))
    return train_dl, test_dl


path = '../dataset/transport.csv'
inputs = 44 # 34 for ionosphere or 44 for transport
epochs = 50

train_dl, test_dl = prepare_data(path, batch_size=128)

model = MLP(inputs)
train_model(train_dl, model, epochs)
acc = evaluate_model(test_dl, model)
print('Accuracy: %.3f' % acc)
```
