## Preparação dos Dados

```python
df = pd.read_csv("dataset/text_classification/bbc-text.csv")
df = df[['text','category']]
num_classes = len(df.category.value_counts())
print('Classes:', num_classes)
df
```

```python
def split_data(df, x_col, y_col, perc_train=0.8):
    len_train = int(len(df)*perc_train)
    dataset_train = df[[x_col, y_col]].iloc[0:len_train, :]
    dataset_test = df[[x_col, y_col]].iloc[len_train:, :]
    print('\nTrain | Test dataset lenght:', len(dataset_train), len(dataset_test))
    return dataset_train, dataset_test

def filter_text(otext, filters):
    text = re.sub(r'(['+filters+'])', r' \1 ', otext)
    text = ' '.join([word for word in text.split() if len(word) > 0 and word != ' '])
    return text
    
def removeSW(otext, stopwords, filters):
    text = otext.lower()
    text = filter_text(text, filters)
    words = [item for item in text.split() if item not in stopwords]
    text = ' '.join(words)
    return text


data = df.copy()
data['otext'] = data.text

filters = "!?'#$%&()*+/:;<=>@[\\]^_`{|}~\t\n+"
swds = stopwords.words('english')
data.text = data.text.apply(lambda x: removeSW(x, swds, filters + ',.'))

dataset_train, dataset_test = split_data(data, x_col='text', y_col='category', perc_train=0.85)
dataset_train
```

```python
encoder = LabelEncoder()
encoder.fit(dataset_train.category)
y_train = keras.utils.to_categorical(encoder.transform(dataset_train.category))
y_test = keras.utils.to_categorical(encoder.transform(dataset_test.category))
print('Shape of targets (train | test):', y_train.shape, y_test.shape)
y_train
```

```python
def get_tokenizer(corpus, num_words, filters):
    oov_token = '' 
    tokenizer = Tokenizer(num_words=num_words, filters=filters)
    #tokenizer = Tokenizer(num_words=num_words, oov_token=oov_token, filters=filters)
    tokenizer.fit_on_texts(corpus)
    vocab_size = len(tokenizer.word_index)
    word2idx = tokenizer.word_index
    idx2word = {v: k for (k, v) in word2idx.items()}
    print('Found %d vocab compared to an allowed %d size' % (vocab_size, num_words))
    if vocab_size > num_words:
        print('Discarding a total of %d less frequent words' % (vocab_size - num_words))
    print('Word counts: %s' % (list(tokenizer.word_counts.items())[:5]))
    print('Most commom words: %s' % (list(tokenizer.word_index.items())[:5]))
    sizes = [len(x) for x in corpus]
    print('Longest sentence (chars+/-std): %d (%d+/-%d)' % (max(sizes), int(np.mean(sizes)), int(np.std(sizes))))
    words = [len(x.split()) for x in corpus]
    print('Longest sentence (words+/-std): %d (%d+/-%d)' % (max(words), int(np.mean(words)), int(np.std(words))))
    return vocab_size, tokenizer, word2idx, idx2word


max_vocab, max_len = 10000, 10

corpus = np.array(dataset_train.text)
vocab_size, tokenizer, word2idx, idx2word = get_tokenizer(corpus, max_vocab, filters + ',.')
```

```python
def use_tokenizer(tokenizer, corpus, printt=False):
    sequences = tokenizer.texts_to_sequences(corpus)
    if printt:
        print('\nTokenized (example):')
        print(corpus[0][:100], '[...]\n-> ', sequences[0])
    return sequences


corpus = np.array(dataset_train.text)
corpus_seqs = use_tokenizer(tokenizer, corpus, printt=True)
```

```python
def align_sentences(sequences, maxlen, padding, truncating):
    return pad_sequences(sequences, maxlen=maxlen, padding=padding, truncating=truncating)

def prepare_dataset(df_train, df_test, max_vocab, filters, punct, max_len, printt=False):
    if not punct: filters = filters + ',.'
    corpus_train, corpus_test = np.array(df_train), np.array(df_test)
    vocab_size, tokenizer, word2idx, idx2word = get_tokenizer(corpus_train, max_vocab, filters)
    train_seqs = use_tokenizer(tokenizer, corpus_train)
    test_seqs = use_tokenizer(tokenizer, corpus_test)
    X_train = align_sentences(train_seqs, max_len, 'post', 'post')
    X_test = align_sentences(test_seqs, max_len, 'post', 'post')
    print('\nDataset shape (train):', X_train.shape)
    if printt: print(X_train[:10])
    print('\nDataset shape (test):', X_test.shape)
    if printt: print(X_test[:10])
    return X_train, X_test, tokenizer, word2idx, idx2word


max_vocab, max_len = 20, 10
X_train, X_test, tokenizer, word2idx, idx2word = prepare_dataset(dataset_train.text, dataset_test.text, 
                                                         max_vocab, filters, False, max_len, printt=True)
```

## Carregamento dos Embeddings Treinados (GloVe)

Nas últimas aulas, apesar de utilizarmos embeddings em todas as tarefas, estávamos os aprendendo do zero baseando-se na base de treinamento.

Na verdade, não há necessidade disso porque já existe uma variedade muito grande de embeddings pré-treinados que podem ser utilizados.

Para este exercício vamos utilizar o [GloVe](https://nlp.stanford.edu/projects/glove/), mas existem muitas opcões de embeddings pré-treinados, inclusive em português.

Antes de rodar as próximas linhas, baixe o arquivo [glove.6B.50d.txt](https://github.com/uclnlp/inferbeddings/blob/master/data/glove/glove.6B.50d.txt.gz)

Vamos ler o conteúdo do arquivo com embeddings pré-treinados, para posteriormente utilizá-los na nossa camada de Embedding. O propósito da próxima célula é determinar o vocabulário para a nossa base de treinamento, e carregar os embeddings para todas as palavras encontradas.

```python
def load_embedding_matrix(tokenizer, max_vocab, embeddings_file, printt=False):
    num_emb = 0
    used_keys = []
    for word in tokenizer.word_index.keys():
        used_keys.append(word)
        num_emb += 1
        if num_emb == max_vocab:
            break

    embeddings_index = dict()
    f = open(embeddings_file)
    for line in f:
        values = line.split()
        word = values[0]
        if word in used_keys:
            coefs = np.array(values[1:], dtype='float32')
            embeddings_index[word] = coefs
    f.close()
    print('\nLoaded %s word vectors' % len(embeddings_index))
    nb_embeddings = len(values)-1
    print('Shape of the loaded embeddings:', nb_embeddings)

    embedding_matrix = np.zeros((max_vocab, nb_embeddings))
    for word, i in tokenizer.word_index.items():
        if i >= max_vocab:
            break
        embedding_vector = embeddings_index.get(word)
        if embedding_vector is not None:
            embedding_matrix[i-1] = embedding_vector

    print('Embeddings matrix shape:', embedding_matrix.shape)
    if printt: print('\nSample:\n', embedding_matrix[0])
    
    return embedding_matrix, nb_embeddings


embeddings_file = '/home/lazarop/Downloads/glove.6B.50d.txt'
embedding_matrix, nb_embeddings = load_embedding_matrix(tokenizer, max_vocab, embeddings_file, printt=True)
```

## Treinamento da rede neural

```python
results = {}
best_model, best_history, best_accuracy, best_key = None, None, 0, 0
start = time.time()

for i, search in enumerate(grid):
    report = ''
    if i > 0: 
        report += 'OVERALL:\nBest accuracy: %.1f%%\n' % (100*best_accuracy)
        report += 'Setup: %s\n' % str(best_model)
        report += '\nETA (hours): %.2f\n' % ((len(grid)-i)*(time.time()-start)/i/60/60)
    report += '\nRunning experiment %d of %d (%.1f%%)\n' % (i+1, len(grid), 100*i/len(grid))
    report += 'Setup: %s\n' % str(search)

    X_train, X_test, tokenizer, word2idx, idx2word = prepare_dataset(dataset_train.text, dataset_test.text, 
                                                         search['vocab'], filters, search['punct'], search['maxlen'])
    k = simplerKeras(X=X_train, y=y_train, gpu=gpu, workers=workers, gpu_test=False)
    
    embedding_matrix, nb_embeddings = load_embedding_matrix(tokenizer, search['vocab'], embeddings_file)

    model = Sequential()
    model.add(Embedding(input_dim=search['vocab'], output_dim=search['embeddings'], weights=[embedding_matrix], 
                        input_length=search['maxlen'], trainable=False))
    if search['bidirectional']:
        model.add(Bidirectional(LSTM(search['lstm'], return_sequences=True, dropout=dropout)))
        model.add(Bidirectional(LSTM(search['lstm'])))
    else:
        model.add(LSTM(search['lstm'], return_sequences=True, dropout=dropout))
        model.add(LSTM(search['lstm']))    
    model.add(Dense(16*num_classes, activation='relu'))
    model.add(Dropout(dropout))
    model.add(Dense(4*num_classes, activation='relu'))
    model.add(Dropout(dropout))
    model.add(Dense(num_classes, activation='softmax'))
    model.compile(loss="categorical_crossentropy", optimizer=optimizers[search['optimizer']], 
                  metrics=["accuracy"])
    model.summary()

    checkpoint = time.time()
    history = model.fit(k.X, k.y, epochs=epochs, batch_size=batch_size, verbose=0, 
                        workers=k.workers, use_multiprocessing=k.multicore,
                        validation_split=0.1,
                        callbacks=[formatCallbackNotebook(epochs,checkpoint,report,timeout)])

    loss, acc = model.evaluate(X_test, y_test, verbose=0)

    results[i] = {'id':i+1, 'vocab': search['vocab'], 'bidirectional': search['bidirectional'],
                  'embeddings': search['embeddings'], 'maxlen': search['maxlen'], 'lstm': search['lstm'], 
                  'punct': search['punct'], 'optimizer': search['optimizer'],
                  'loss': loss, 'acc': acc, 'epochs': len(history.history['loss']),
                  'elapsed_time': time.time()-checkpoint}

    if acc > best_accuracy:
        best_model = search
        best_history = history
        best_accuracy = acc
        best_key = i
    if i + 1 < len(grid):
        del model
        k.release()
        del k
```

```python
def printQuadrants(results, epochs, timeout, results_file, min_acc=0.0):
    print('\nResults:')
    print('Training restrictions: %.0f max minutes and %d max epochs per training' % (timeout, epochs))
    aux = pd.DataFrame.from_dict(results, orient='index')
    aux['timeout'] = timeout
    aux['max_epochs'] = epochs
    try:
        df = pd.read_csv(results_file)
    except:
        df = pd.DataFrame()
    df = pd.concat([df, aux], axis=0)
    for col in ['acc','loss','elapsed_time']:
        df[col] = df[col].apply(lambda x: round(x, 4)) 
    df = df.drop_duplicates().reset_index(drop=True)
    df.sort_values(by='acc', inplace=True, ascending=False)
    df.to_csv(results_file, index=False)
    df = df.loc[df.acc >= min_acc]

    cols_analyse = [cols for cols in df.columns if not cols in [x] + [y] + ign]
    remarks, graphs = [], []
    for col in cols_analyse:
        if len(df[col].unique()) > 1:
            graphs.append([x, y, col])
        else:
            remarks.append('%s (%s)' % (col, str(df[col].iloc[0])))

    if len(remarks) > 0: print('Constant parameters:', ', '.join(remarks))

    plt.rcParams.update({'font.size': 14})
    fig, axes = plt.subplots(figsize=(15, 8*len(graphs)), nrows=len(graphs))    
    for i, cols in enumerate(graphs):
        aux = df[cols].copy()
        names = aux.columns
        if len(aux[names[2]].unique()) <= 10:
            sns.scatterplot(x=names[0] ,y=names[1], hue=names[2], data=aux, alpha=0.5, 
                            ax=axes[i], s=300, legend='full')  
        else:
            sns.scatterplot(x=names[0] ,y=names[1], size=names[2], data=aux, alpha=0.5, 
                            ax=axes[i], s=300)  
        axes[i].tick_params(axis='both', labelsize=12)
        axes[i].tick_params(axis='both', labelsize=12)
        axes[i].set_title(('Analyse: %s' % names[2]).upper(), size=16)
        axes[i].legend(fontsize=16)
    plt.plot()
    display(df)
    
def printHistory(history, metric='accuracy'):
    fig, axs = plt.subplots(nrows=1, ncols=2, figsize = (14, 7))
    axs[0].plot(history[metric], label='Train')
    axs[0].plot(history['val_'+metric], label='Train')
    axs[0].set_xlabel('Epoch')
    axs[0].set_title('Model Performance')
    axs[0].legend()
    axs[1].plot(history['loss'], label='Train')
    axs[1].plot(history['val_loss'], label='Train')
    axs[1].set_xlabel('Epoch')
    axs[1].set_title('Model Loss')
    axs[1].legend()
    plt.show() 
    
    
x, y, ign = 'loss', 'acc', ['id', 'loss']

printQuadrants(results, epochs, timeout, 'dataset/results_bbc.csv', min_acc=0.80)
    
printHistory(best_history.history, metric='accuracy')
```

## Tradutor (seq2seq)

Agora que já somos experts em classificar sentimentos em textos, vamos considerar uma tarefa diferente mas igualmente importante: **Tradução**.

Como realizar traducões entre duas línguas requer modelos grandes e um grande corpus de treinamento, vamos considerar uma tarefa de "tradução" mais simples.

A tarefa que consideraremos é a traducão de uma data em texto livre para um formato mais palatável para o computador. Alguns exemplos de traducão estão abaixo:

`9 may 1998 -> 1998-05-09
10.11.19 -> 2019-11-10
9/10/70 -> 1970-09-10
saturday april 28 1990 -> 1990-04-28
thursday january 26 1995 -> 1995-01-26
monday march 7 1983 -> 1983-03-07`

Para realizar esta tarefa, modelaremos o problema de uma forma um pouco diferente. Para a criação de nosso vocabulário, designaremos um código para cada caractere (não cada palavra), formando um vocabulario de origem e um de destino.

### The neural translation model is an encoder-decoder recurrent neural network

**Etapas:**
* 1. Selecionar pares de sequências (input_sequences) de um domínio para sua correspondência (target_sequences) em outro domínio (ex. idioma A para B)
* 2. Desenvolver um encoder LSTM que transforma as input_sequences em 2 state vectors (mantemos o último estado e descartamos as saídas)
* 3. Desenvolver um decoder LSTM que transforma as target_sequences na mesma sequência, porém deslocado 1 passo no futuro (teacher forcing)
    * este decoder usa como initial state o state vector do encoder acima
    * o decoder aprende a gerar targets(t+1...) a partir de targets(...t), condicionado às input_sequences
* 4. Ao realizar predições, queremos decodificar uma input_sequence desconhecida
    * encode a sequência de entrada em state vectors
    * iniciar com uma target_sequence de tamanho 1
    * alimentar os state vectors e sequências de tamanho 1 no decoder para predizer o próximo caracter
    * amostrar o próximo caracter usando argmax
    * adicionar o novo caracter amostrado à target_sequence
    * repetir este processo até que seja gerado o caracter de end-of-sequence ou o limite de caracteres


```python
# adaptado de https://github.com/keras-team/keras/blob/master/examples/lstm_seq2seq.py

batch_size = 64 
epochs = 100
latent_dim = 256  # dimensionalidade latente do encoding space


# preparação dos dados

print('Reading CSV database')
data = pd.read_csv('dataset/translation/dates_dataset.csv', header=None)
data.columns = ['input_texts','target_texts']
data.target_texts = data.target_texts.apply(lambda x: '\t' + x + '\n') # marcação de início e término de sequência
train_index = int(0.90 * len(data))
train_data = data[:train_index]
test_data = data[train_index:]
print('Train dataset shape:', train_data.shape)


# vetorização

input_texts, target_texts = [], []
input_characters, target_characters = set(), set()
for _, row in train_data.iterrows():
    input_text, target_text = row['input_texts'], row['target_texts']
    input_texts.append(input_text)
    target_texts.append(target_text)
    for char in input_text:
        if char not in input_characters:
            input_characters.add(char)
    for char in target_text:
        if char not in target_characters:
            target_characters.add(char)

print('Sample of input -> target:', input_texts[0], '->', target_texts[0][1:]) # início em 1 para não imprimir o TAB


# tokenização

input_characters = sorted(list(input_characters))
target_characters = sorted(list(target_characters))
num_encoder_tokens = len(input_characters)
num_decoder_tokens = len(target_characters)
max_encoder_seq_length = max([len(txt) for txt in input_texts])
max_decoder_seq_length = max([len(txt) for txt in target_texts])

print('Number of samples:', len(input_texts))
print('Number of unique input tokens:', num_encoder_tokens)
print('Number of unique output tokens:', num_decoder_tokens)
print('Max sequence length for inputs:', max_encoder_seq_length)
print('Max sequence length for outputs:', max_decoder_seq_length)

input_token_index = dict([(char, i) for i, char in enumerate(input_characters)])
target_token_index = dict([(char, i) for i, char in enumerate(target_characters)])
print('\nInput tokens:', input_token_index)
print('Target tokens:', target_token_index)

reverse_input_char_index = dict((i, char) for char, i in input_token_index.items())
reverse_target_char_index = dict((i, char) for char, i in target_token_index.items())
print('\nReverse input tokens:', reverse_input_char_index)
print('Reverse target tokens:', reverse_target_char_index)


# inicialização dos encoders / decoders

encoder_input_data = np.zeros((len(input_texts), max_encoder_seq_length, num_encoder_tokens), dtype='float32')
decoder_input_data = np.zeros((len(input_texts), max_decoder_seq_length, num_decoder_tokens), dtype='float32')
decoder_target_data = np.zeros((len(input_texts), max_decoder_seq_length, num_decoder_tokens), dtype='float32')

print('\nShape of encoder_input_data:', encoder_input_data.shape)
print('Shape of decoder_input_data:', decoder_input_data.shape)
print('Shape of decoder_target_data:', decoder_target_data.shape)

for i, (input_text, target_text) in enumerate(zip(input_texts, target_texts)):
    for t, char in enumerate(input_text):
        encoder_input_data[i, t, input_token_index[char]] = 1.
    encoder_input_data[i, t + 1:, input_token_index[' ']] = 1. # padding post com espaços em branco
    for t, char in enumerate(target_text):
        # decoder_target_data is ahead of decoder_input_data by one timestep
        decoder_input_data[i, t, target_token_index[char]] = 1.
        if t > 0:
            decoder_target_data[i, t - 1, target_token_index[char]] = 1.
    decoder_input_data[i, t + 1:, target_token_index['\n']] = 1. # padding post com espaços em branco
    decoder_target_data[i, t:, target_token_index['\n']] = 1. # padding post com espaços em branco


# modelagem da topologia do encoder e do decoder

print('\nSetup for the encoder')
encoder_inputs = Input(shape=(None, num_encoder_tokens))
print('Shape da layer encoder_inputs:', encoder_inputs.shape)
encoder = LSTM(latent_dim, return_state=True)
_, state_h, state_c = encoder(encoder_inputs) # a saída desta camada é descartada (encoder_outputs)
print('Shape do state_h:', state_h.shape)
print('Shape do state_h:', state_h.shape)
encoder_states = [state_h, state_c] 

print('\nSetup for the decoder')
decoder_inputs = Input(shape=(None, num_decoder_tokens)) # os estados do encoder são usados como initial state
print('Shape da layer decoder_inputs:', decoder_inputs.shape)
# os estados serão usados posteriormente para fazer predições
decoder_lstm = LSTM(latent_dim, return_sequences=True, return_state=True)
decoder_outputs, _, _ = decoder_lstm(decoder_inputs, initial_state=encoder_states)
print('Shape da layer decoder_outputs:', decoder_outputs.shape)

decoder_dense = Dense(num_decoder_tokens, activation='softmax')
decoder_outputs = decoder_dense(decoder_outputs)
print('Shape da layer decoder_outputs (pós dense):', decoder_outputs.shape)

model = Model([encoder_inputs, decoder_inputs], decoder_outputs)

print('\nPlease wait (training model to file s2s.h5)')
model.compile(optimizer='rmsprop', loss='categorical_crossentropy', metrics=['accuracy'])
model.fit([encoder_input_data, decoder_input_data], decoder_target_data,
          batch_size=batch_size, epochs=epochs, validation_split=0.2, verbose=0)
# Save model
model.save('s2s.h5')


# modelagem da topologia para realizar inferências

print('\nSetup for inference mode')
encoder_model = Model(encoder_inputs, encoder_states)

decoder_state_input_h = Input(shape=(latent_dim,))
decoder_state_input_c = Input(shape=(latent_dim,))
print('Shape da layer decoder_state_input_h:', decoder_state_input_h.shape)
print('Shape da layer decoder_state_input_c:', decoder_state_input_c.shape)
decoder_states_inputs = [decoder_state_input_h, decoder_state_input_c]

decoder_outputs, state_h, state_c = decoder_lstm(decoder_inputs, initial_state=decoder_states_inputs)
print('Shape da layer decoder_outputs (inference mode):', decoder_outputs.shape)
decoder_states = [state_h, state_c]

decoder_outputs = decoder_dense(decoder_outputs)
print('Shape da layer decoder_outputs (pós dense):', decoder_outputs.shape)

decoder_model = Model([decoder_inputs] + decoder_states_inputs, [decoder_outputs] + decoder_states)


# função para gerar a sequência de saída

def decode_sequence(input_seq):
    states_value = encoder_model.predict(input_seq) # codificar a input como um vetor de estados
    target_seq = np.zeros((1, 1, num_decoder_tokens))
    target_seq[0, 0, target_token_index['\t']] = 1. # popular o primeiro caracter da sequência pelo padrão
    stop_condition = False
    decoded_sentence = ''
    while not stop_condition:
        output_tokens, h, c = decoder_model.predict([target_seq] + states_value)
        sampled_token_index = np.argmax(output_tokens[0, -1, :])
        sampled_char = reverse_target_char_index[sampled_token_index]
        decoded_sentence += sampled_char
        if (sampled_char == '\n' or len(decoded_sentence) > max_decoder_seq_length):
            stop_condition = True
        target_seq = np.zeros((1, 1, num_decoder_tokens))
        target_seq[0, 0, sampled_token_index] = 1.
        states_value = [h, c]
    return decoded_sentence
```

# Gerador de nomes de dinossauros

Para o último problema a ser resolvido, vamos considerar um problema um pouco mais exótico, semelhante ao problema de geracão de palavras estudado na segunda aula.

Para esta tarefa geraremos **nomes de Dinossauros**.

Nomes de dinossauros têm uma certa semelhança, por exemplo:

`
Abdallahsaurus
Abelisaurus
Abrictosaurus
`

Portanto, vamos treinar um modelo que completa o nome de um dinossauro!

Para isto, novamente vamos gerar um token para cara caractere. O modelo deve ser treinado recebendo um nome parcial e o completando, por exemplo:

`
x = 'Abda', y = 'Abdallahsaurus'
x = 'Abelisa', y = 'Abelisaurus'
`

O modelo pode ser treinado com a base de treinamento sendo gerada similarmente ao problema de gerar sonetos, mas adicionando um token "\n" à saída para indicar o final do nome do Dinossauro:

`
x = 'A', y = 'Aardonyx\n'
x = 'Aa', y = 'Aardonyx\n'
x = 'Aar', y = 'Aardonyx\n'
...
x = 'Aardony', y = 'Aardonyx\n'
`

### Preparação dos dados

```python
def clear_string(otext, filters):
    text = otext.replace('...',',')
    text = re.sub(r'(['+filters+'])', r'', text) # remover qualquer caracter inválido indicado em filters
    return text.lower()

def get_data(frac_train=0.8, printt=False, filters=''):
    row_list = []
    largest_name = 0
    with open("dataset/text_generation/dinos.txt", "r") as a_file:
        for line in a_file:
            largest_name = max(largest_name, len(line))
            if line[-1] != '\n': line += '\n' # adicionar o caracter de fim de string caso não exista
            for i in range(1, len(line)): # vamos iniciar com sequências de tamanho mínimo 2
                row_list.append([line, line[:i+1]])

    df = pd.DataFrame(row_list, columns=['dino','seqs'])
    df.seqs = df.seqs.apply(lambda x: clear_string(x, filters))
    data = df.copy().sample(frac=1).reset_index(drop=True)
    training_index = int(frac_train * len(data))
    train_data = data.loc[:training_index, :].reset_index(drop=True)
    test_data = data.loc[training_index:, :].reset_index(drop=True)
    if printt:
        print('\nShape of seqs dataset:', df.shape)
        print(df)
        print('Shape of train/test dataset:', train_data.shape, test_data.shape)
    return train_data, test_data, df


filters = "!?'#$%&()*+/:;<=>@[\\]^_`{|}~\t0123456789"
train_data, test_data, _ = get_data(0.85, printt=True, filters=filters)
```

```python
def align_sentences(sequences, maxlen, padding, truncating):
    return pad_sequences(sequences, maxlen=maxlen+1, padding=padding, truncating=truncating)

def prepare_data(frac_train=0.85, printt=True, filters=filters):
    # coletar dados e separar base de treino e teste
    train_data, test_data, df = get_data(frac_train, printt, filters)
    # preparar e realizar a vetorização dos dados por caracter
    tokenizer = Tokenizer(char_level=True)
    tokenizer.fit_on_texts(train_data.loc[:, 'seqs'])
    word2idx = tokenizer.word_index
    idx2word = {v: k for k, v in word2idx.items()}
    # preparar representação dos dados em formato matricial
    seq_train = tokenizer.texts_to_sequences(train_data.loc[:, 'seqs'])
    seq_test = tokenizer.texts_to_sequences(test_data.loc[:, 'seqs'])
    vocab_size = len(word2idx)
    max_len = max([len(seq)-1 for seq in seq_train] + [len(seq)-1 for seq in seq_test])
    seq_train = align_sentences(seq_train, max_len, 'pre', 'pre')
    seq_test = align_sentences(seq_test, max_len, 'pre', 'pre')
    # separação das variáveis X e y
    X_train = np.array(seq_train[:, :-1])
    y_train = keras.utils.to_categorical(seq_train[:, -1], num_classes=vocab_size+1)
    X_test = np.array(seq_test[:, :-1])
    y_test = keras.utils.to_categorical(seq_test[:, -1], num_classes=vocab_size+1)
    if printt:
        print('\nTotal vocab size:', vocab_size)
        print('Max length:', max_len)
        print('\nword2Idx:', word2idx)
        print('\nidx2Word:', idx2word)
        print('\nSeqs sample (train):', train_data.loc[0, 'seqs'], '->', seq_train[0])
        print('Seqs sample (test):', test_data.loc[0, 'seqs'], '->', seq_test[0])
        print('\nDataset shape X_train | y_train | X_test | y_test:', X_train.shape, y_train.shape, X_test.shape, y_test.shape)    
        print('\nTrain sample:\n', X_train[0], '\n->', y_train[0])
        print('\nTest sample:\n', X_test[0], '\n->', y_test[0])
    return X_train, y_train, X_test, y_test, tokenizer, vocab_size, max_len, word2idx, idx2word

     
printt = True
frac_train = 0.85
filters = "!?'#$%&()*+/:;<=>@[\\]^_`{|}~\t0123456789"
X_train, y_train, X_test, y_test, tokenizer, vocab_size, max_len, word2idx, idx2word = prepare_data(
                                              frac_train=frac_train, printt=printt, filters=filters)
```

### Treinamento da rede neural

```python
# preparar os dados
printt = False
frac_train = 0.85
filters = "!?'#$%&()*+/:;<=>@[\\]^_`{|}~\t0123456789"
X_train, y_train, X_test, y_test, tokenizer, vocab_size, max_len, word2idx, idx2word = prepare_data(frac_train=frac_train, printt=printt, filters=filters)

# configurar o treinamento
gpu = True
workers = 10
batch_size, epochs, dropout, timeout = 64, 200, 0.1, 1


optimizers = {'adam': Adam(lr=1e-3, decay=1e-3/epochs), 'adam_fast': Adam(lr=1e-1, decay=1e-1/epochs), 
              'rmsprop': RMSprop(lr=1e-3, rho=0.9, momentum=0.0), 'rmsprop_fast': RMSprop(lr=1e-1, rho=0.9, momentum=0.0),
              'sgd': SGD(lr=1e-3, momentum=0.9), 'sgd_fast': SGD(lr=1e-1, momentum=0.9)}

models = {'RNN1': RecurrentOne, 'RNN2': RecurrentTwo,
          'RNN3': RecurrentThree, 'RNN4': RecurrentFour}

grid = []
for vs in [vocab_size+1]: # tokens iniciam em 1, adicionar posição do zero
    for ed in [8]:
        for punct in [True]:
            for opt in ['adam','adam_fast','sgd','sgd_fast','rmsprop','rmsprop_fast']:
            #for opt in ['adam']:
                for lstm in [max_len]:
                    for ml in [max_len]:
                        for m in ['RNN4', 'RNN3', 'RNN2', 'RNN1']:
                        #for m in ['RNN2']:
                            grid.append({'model': m, 'vocab': vs, 'embeddings': ed, 'maxlen': ml,
                                         'lstm': lstm, 'punct': punct, 'optimizer': opt})
                            
print('Up to %.2f hours to complete those %d trainings' % ((len(grid)*timeout/60), len(grid)))
```

```python
# análise dimensional da topologia (TESTE)

inputs = Input(shape=(max_len,), name='inputs')
print('Shape da entrada:', inputs.shape)
# Batch size x Comprimento de cada X (X.shape[1])

x = Embedding(input_dim=vocab_size+1, output_dim=8, input_length=max_len, name='embeddings')(inputs)
print('Shape da layer embedding:', x.shape)
# Batch size x Timesteps (length) x Features (embeddings dimensions)

x, state_h, state_c = LSTM(max_len, return_state=True, name='lstm')(x)
print('Shape da layer LSTM:', x.shape)
print('Shape dos estados da layer LSTM:', state_h.shape)
# Batch size x Células de memória

x = Dense(y_train.shape[1], activation='softmax', name='dense')(x)
print('Shape da saída:', x.shape)
# Batch size x Comprimento de cada y (y.shape[1])

model = Model([inputs], x)

model.compile(loss="categorical_crossentropy", optimizer=optimizers['rmsprop'], metrics=["accuracy"])
model.summary()

history = model.fit(X_train, y_train, epochs=10, batch_size=batch_size, verbose=2, validation_split=0.1)
```

```python
results = {}
best_model, best_history, best_accuracy, best_key = None, None, 0, 0
start = time.time()

for i, search in enumerate(grid):
    report = ''
    if i > 0: 
        report += 'OVERALL:\nBest accuracy: %.1f%%\n' % (100*best_accuracy)
        report += 'Setup: %s\n' % str(best_model)
        report += '\nETA (hours): %.2f\n' % ((len(grid)-i)*(time.time()-start)/i/60/60)
    report += '\nRunning experiment %d of %d (%.1f%%)\n' % (i+1, len(grid), 100*i/len(grid))
    report += 'Setup: %s\n' % str(search)
    
    k = simplerKeras(X=X_train, y=y_train, gpu=gpu, workers=workers, gpu_test=False)

    model = models[search['model']].build(input_dim=search['vocab'], output_dim=search['embeddings'], 
                                  input_len=search['maxlen'], lstm_out=search['lstm'], dropout=dropout,
                                  classes=y_train.shape[1], reg=l2(0.0005), last_act='softmax')
    model.compile(loss="categorical_crossentropy", optimizer=optimizers[search['optimizer']], metrics=["accuracy"])
    model.summary()

    checkpoint = time.time()
    history = model.fit(k.X, k.y, epochs=epochs, batch_size=batch_size, verbose=0, 
                        workers=k.workers, use_multiprocessing=k.multicore, validation_split=0.1, 
                        callbacks=[formatCallbackNotebook(epochs,checkpoint,report,timeout)])

    loss, acc = model.evaluate(X_test, y_test, verbose=0)
        
    results[i] = {'id':i+1, 'model': search['model'], 'vocab': search['vocab'], 
                  'embeddings': search['embeddings'], 'maxlen': search['maxlen'], 'lstm': search['lstm'], 
                  'punct': search['punct'], 'optimizer': search['optimizer'],
                  'loss': loss, 'acc': acc, 'epochs': len(history.history['loss']),
                  'elapsed_time': time.time()-checkpoint}

    if acc > best_accuracy:
        best_model = search
        best_history = history
        best_accuracy = acc
        best_key = i
    if i + 1 < len(grid):
        del model
        k.release()
        del k
```

```python
def printQuadrants(results, epochs, timeout, results_file, min_acc=0.0):
    print('\nResults:')
    print('Training restrictions: %.0f max minutes and %d max epochs per training' % (timeout, epochs))
    aux = pd.DataFrame.from_dict(results, orient='index')
    aux['timeout'] = timeout
    aux['max_epochs'] = epochs
    try:
        df = pd.read_csv(results_file)
    except:
        df = pd.DataFrame()
    df = pd.concat([df, aux], axis=0)
    for col in ['acc','loss','elapsed_time']:
        df[col] = df[col].apply(lambda x: round(x, 4)) 
    df = df.drop_duplicates().reset_index(drop=True)
    df.sort_values(by='acc', inplace=True, ascending=False)
    df.to_csv(results_file, index=False)
    df = df.loc[df.acc >= min_acc]

    cols_analyse = [cols for cols in df.columns if not cols in [x] + [y] + ign]
    remarks, graphs = [], []
    for col in cols_analyse:
        if len(df[col].unique()) > 1:
            graphs.append([x, y, col])
        else:
            remarks.append('%s (%s)' % (col, str(df[col].iloc[0])))

    if len(remarks) > 0: print('Constant parameters:', ', '.join(remarks))

    plt.rcParams.update({'font.size': 14})
    fig, axes = plt.subplots(figsize=(15, 8*len(graphs)), nrows=len(graphs))    
    for i, cols in enumerate(graphs):
        aux = df[cols].copy()
        names = aux.columns
        if len(aux[names[2]].unique()) <= 10:
            sns.scatterplot(x=names[0] ,y=names[1], hue=names[2], data=aux, alpha=0.5, 
                            ax=axes[i], s=300, legend='full')  
        else:
            sns.scatterplot(x=names[0] ,y=names[1], size=names[2], data=aux, alpha=0.5, 
                            ax=axes[i], s=300)  
        axes[i].tick_params(axis='both', labelsize=12)
        axes[i].tick_params(axis='both', labelsize=12)
        axes[i].set_title(('Analyse: %s' % names[2]).upper(), size=16)
        axes[i].legend(fontsize=16)
    plt.plot()
    display(df)
    
def printHistory(history, metric='accuracy'):
    fig, axs = plt.subplots(nrows=1, ncols=2, figsize = (14, 7))
    axs[0].plot(history[metric], label='Train')
    axs[0].plot(history['val_'+metric], label='Train')
    axs[0].set_xlabel('Epoch')
    axs[0].set_title('Model Performance')
    axs[0].legend()
    axs[1].plot(history['loss'], label='Train')
    axs[1].plot(history['val_loss'], label='Train')
    axs[1].set_xlabel('Epoch')
    axs[1].set_title('Model Loss')
    axs[1].legend()
    plt.show() 
    
    
x, y, ign = 'loss', 'acc', ['id', 'loss']

printQuadrants(results, epochs, timeout, 'dataset/results_dinos.csv', min_acc=0.00)
    
printHistory(best_history.history, metric='accuracy')
```

```python
def select_predicted(probs):
    rnd = np.random.random() * np.sum(probs)
    i, cumsum = 0, 0
    while cumsum <= rnd:
        cumsum += probs[i]
        i += 1
    return i-1

def define_weights(size, cap=0.1, force=100):
    w = [1-cap*cap**(i/force) for i in range(size)]
    return w
    
def get_next_char(model, seed, tokenizer, idx2word, weigths, length):
    sequences = tokenizer.texts_to_sequences([seed])
    padded = pad_sequences(sequences, maxlen=length , padding='pre', truncating='pre')
    probs = model.predict(padded).flatten() * weigths
    predicted = select_predicted(probs)
    return str(idx2word[predicted]) #output_word
    
weigths = define_weights(k.y.shape[1], cap=0.01, force=5)
plt.plot(weigths)
plt.title('Custom Weigths for Predict')
plt.ylim(0, 1.1)
plt.plot()

get_next_char(model, 'test', tokenizer, idx2word, weigths, max_len)  
```
