# Problema: prever o intervalo de tempo para colheita

# Resultados Alcançados
previsões feitas a 86 dias da colheita (+/- 29 dias) - dados até estádio 5  

* R2 na base de teste de 86%
* RMSE de 12 dias na base de treino e 10 dias na base de teste

## Fatos Relevantes
* até 3 camadas ocultas do tipo dense
* 1777 parâmetros treináveis no modelo vencedor
* testados 10.000 modelos em 120 minutos (83 testes/minuto) usando 10 workers
* tecnologias: MLP ANN + gridsearch + algoritmo genético em 20 gerações
* 16 genes mutáveis no DNA, mantendo cepa de 40% fazendo crossover de 30% e mutação de 30%

# Passo a Passo - Solução
* importar todos os métodos de rubia_ann
    * neste arquivo criei uma série de classes e métodos para abstrair as redes
    * consiste de um algoritmo genérico associado ao GridSearchCV e processos de threading
* carregar os dados do CSV
    * prever delta sendo conhecida a evolução das culturas até E5
* separar os dados em conjunto de treino/teste e conjunto de validação
    * como existe auto correlação (dependência temporal) dentro da uma mesma safra, vamos usar 80% das safras para treino e 20% para validação, sem misturar
* fazer o pre processamento dos dados, especialmente ajuste de dimensão em X
    * X: shape (463,42) sendo 463 observações e 42 variáveis preditoras
    * y: shape (463,1) sendo 463 observações da variável dependente "delta"
* inicializar os parâmetros necessários para uso das classes
    * input_dim e output_dim: valores utilizados pelas 2 camadas visíveis (entrada e saída)
    * mode: este é um problema de regressão com 42 entradas e 1 saída contínua
    * topology: topologia máxima da rede a ser construída, excluindo a layer de saída
        * neste exercício vamos testar diferentes topologias, contendo até 3+1 Dense
    * metric: vamos usar o MSE pois é adequado a este tipo de problema, reavaliamos se necessário
    * GENES: aqui incluímos as mutações permitidas em cada um dos tipos de genes
        * para manter um gene imutável, informar uma lista com este único valor permitido
    * também vamos usar um timer para medir o custo total dessa busca por hiperparâmetros
* inicializar a classe simpler_keras
    * X e y: numpy arrays com os dados de treino/teste
    * para esse conjunto onde dimX é pequena, temos melhor desempenho usando múltiplos workers na CPU do que 1 GPU
    * vamos usar todos os cores da CPU exceto 2 (n_jobs=-3)
* checamos se tudo foi inicializado corretamente dand um check no objeto criado
* inicializamos a primeira cepa de indivíduos por meio do método setGenetic
    * informamos a topologia, o tamanho da população e o número de gerações a ser modeladas
    * definimos um percentual da população que será separado para criar a nova geração de indivíduos
    * os melhores indivíduos (com mais alto scores) são selecionados
* executamos a busca pelos melhores parâmetros
    * definimos número de folds e percentual da amostra usados na validação durante o treinamento
    * os melhores indivíduos são mantidos, os demais são clonados destes pais
    * se um novo indivíduo não for gerado via cruzamento, ele será gerado via mutação
    * definimos o percentual de indivíduos que serão gerados a partir do cruzamento de pais selecionados
    * definimos o número de genes que devem sofrer mutação para a próxima geração
    * adicionamos um handcap (%) para penalizar modelos com alta variabilidade (instabilidade) ou computacionalmente caro (tempo)
    * o handcap é descontado do score de cada indivíduo, sendo o score a média entre treino e validação
    * inicializamos o timer e exibimos demais estatísticas
    * o melhor modelo é salvo em myobj.best_model, podendo ser usado posteriormente
* finalmente, testamos o modelo usando os dados de teste reservados em X_test e y_test
    * avaliamos diversas métricas para conferir a qualidade e robustez do modelo selecionado

```python
import pandas as pd
import numpy as np
import random

from rubia_ann import *
```

```python
data = pd.read_csv('dataset/e5.csv')
print(data.shape)
data.sample(7).T
```

```python
def checkData(data, col):
    print(col, len(data.loc[data[col].isnull()]))

for col in data.columns:
    checkData(data, col)
    
data.info()
```

```python
print(data.delta.mean())
print(data.delta.std())
```

```python
data_train, data_test = pd.DataFrame(), pd.DataFrame()
train_perc = 0.8

for codcult in data.codcult.unique():
    subset = data.loc[data.codcult==codcult]
    safras = list(subset.safra.unique())
    saf_train = random.sample(safras, int(train_perc * len(safras)))
    data_train = pd.concat([data_train, subset.loc[subset.safra.isin(saf_train)]], axis=0)
    data_test = pd.concat([data_test, subset.loc[~subset.safra.isin(saf_train)]], axis=0)

print('Train', data_train.groupby(['codcult','safra']).count()['delta'])
print('Test', data_test.groupby(['codcult','safra']).count()['delta'])

data_train.drop('safra', axis=1, inplace=True)
data_test.drop('safra', axis=1, inplace=True)
```

```python
data_train = np.array(data_train)
X_train, y_train = data_train[:,:-1], data_train[:,-1]
scalarX = MinMaxScaler()
scalarX.fit(X_train)
X_train = scalarX.transform(X_train)
X_train[:3]
```

```python
data_test = np.array(data_test)
X_test, y_test = data_test[:,:-1], data_test[:,-1]
X_test = scalarX.transform(X_test)
X_test[:3]
```

```python
print('Shape train', data_train.shape, 'Shape test', data_test.shape)
```

```python
input_dim = 42
output_dim = 1
mode = 'regression'

topology = ['Dense','Dense','Dense']
metric = ['mse']

GENES = {
    'optimizers': ['rmsprop','adam'],
    'losses': ['mse'],
    'activators': ['relu','elu','linear'],
    'last_activation': ['relu','linear'],
    'denses': [0] + [2 ** i for i in range(3, 7)],
    'dropout': [True, False],
    'dropout_rate': [0.0, 0.1, 0.3, 0.5],
    'epochs': [50, 100, 200],
    'batch_size': [10, 50, 200, 400]
}
```

```python
k = simpler_keras(X=X_train, y=y_train, mode=mode, gpu=False, workers=-3, fixed_genes=GENES)
```

```python
k.check()
```

```python
k.setGenetic(topology=topology, population=100, generations=20, keep_portion=0.4, 
             input_dim=input_dim, output_dim=output_dim, metrics=metric)
```

```python
start = time.time()

k.runGenerations(cv=5, validation_split=0.2, n_mutations=2, crossover=0.5, 
                 performance_cap=0.1, verbose=0)

print('Total elapsed time (s): %.2f' % (time.time() - start))
```

```python
k.printPerformanceR(X_test, y_test)
```
