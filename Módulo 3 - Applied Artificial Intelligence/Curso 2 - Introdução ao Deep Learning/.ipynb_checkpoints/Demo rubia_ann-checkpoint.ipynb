```python
import pandas as pd
import numpy as np
import seaborn as sns
from scipy import stats

from rubia_ann import *
```

# Classificação Binária

```python
data = pd.read_csv('https://raw.githubusercontent.com/jbrownlee/Datasets/master/pima-indians-diabetes.csv', header=None)
print(data.sample(10))
data[8].value_counts()
```

```python
# balanceamento das classes via sampling
data = pd.concat([data.loc[data[8]==0].sample(268), data.loc[data[8]==1]], axis=0)
print(data.sample(10))
data[8].value_counts()
```

```python
data = np.array(data)
X, y = data[:,:-1], data[:,-1]
scalarX, scalarY = MinMaxScaler(), MinMaxScaler()
scalarX.fit(X)
X = scalarX.transform(X)
X
```

## Exemplo Minimalista
Todos os valores de parâmetros opcionais usando valores default

```python
input_dim = 8
output_dim = 1
mode = 'binary'

k = simpler_keras(X=X, y=y, mode=mode)

k.check()

k.setGenetic(input_dim=input_dim, output_dim=output_dim)

k.runGenerations()
```

## Exemplo Robusto
Parâmetros personalizados

```python
input_dim = 8
output_dim = 1
mode = 'binary'
topology = ['Dense','Dense','Dense']
metric = ['accuracy']

GENES = {
    'optimizers': ['adam'],
    'losses': ['binary_crossentropy'],
    'activators': ['relu'],
    'last_activation': ['sigmoid','softmax'],
    'denses': [0]+[2 ** i for i in range(2, 5)],
    'dropout': [True, False],
    'dropout_rate': [0.1, 0.3, 0.5],
    'epochs': [50, 100, 150],
    'batch_size': [50, 200, 800]
}

# gpu and fixed_genes below are optional parameters 
start = time.time()

k = simpler_keras(X=X, y=y, mode=mode, gpu=False, workers=-3, fixed_genes=GENES)

k.check()

k.setGenetic(topology=topology, population=100, generations=20, keep_portion=0.4, 
             input_dim=input_dim, output_dim=output_dim, metrics=metric)

k.runGenerations(cv=5, validation_split=0.1, n_mutations=2, crossover=0.5, 
                 performance_cap=0.2, verbose=0)

print('Total elapsed time (s): %.2f' % (time.time() - start))
```

```python
data = pd.read_csv('https://raw.githubusercontent.com/advinstai/Aprendizagem-estatistica-regressao/master/exercicios/Advertising.csv')
data.drop('Unnamed: 0', axis=1, inplace=True)
data.sample(10)
```

```python
# antes de aplicar Box-Cox, precisamos verificar se existem valores nulos/negativos e pre processar se sim

for col in ['TV','radio','newspaper']:
    print('Minimum for %s:'%col, data[col].min())
```

```python
# somando um valor para evitar 
data.loc[data.radio==0, 'radio'] = 0.001
```

```python
# observamos que as variáveis de interesse apresentam distribuição não normal
#   vamos aplicar transformação de Box-Cox à coluna com skewness e uma quadrática para as 2 uniformes

for col in ['TV','radio','newspaper']:
    x = data[col]
    fig = plt.figure()
    ax1 = fig.add_subplot(211)    
    prob = stats.probplot(x, dist=stats.norm, plot=ax1)
    ax1.set_xlabel('')
    ax1.set_title('Probplot against normal distribution')
    ax2 = fig.add_subplot(212)
    data[col], coefs = stats.boxcox(x)
    prob = stats.probplot(data[col], dist=stats.norm, plot=ax2)
    ax2.set_title('Probplot after Box-Cox transformation')
    plt.show()
    print('Box-Cox lambda (%s):'%col, coefs)
```

```python
sns.pairplot(data)
plt.suptitle('Correlation Pairplot', size=16)
plt.subplots_adjust(top=0.95)
plt.show()
```

```python
npdata = np.array(data)
X, y = npdata[:,:-1], npdata[:,-1]
scalarX, scalarY = MinMaxScaler(), MinMaxScaler()
scalarX.fit(X)
X = scalarX.transform(X)
X[:10]
```

```python
input_dim = 3
output_dim = 1
mode = 'regression'

topology = ['Dense','Dense']
metric = ['mse']

GENES = {
    'optimizers': ['rmsprop','adam'],
    'losses': ['mse'],
    'activators': ['relu'],
    'last_activation': ['relu','linear'],
    'denses': [2 ** i for i in range(3, 5)],
    'dropout': [False],
    'dropout_rate': [0.0],
    'epochs': [200],
    'batch_size': [20]
}

# gpu and fixed_genes below are optional parameters 
start = time.time()

k = simpler_keras(X=X, y=y, mode=mode, gpu=False, workers=-3, fixed_genes=GENES)

k.check()

k.setGenetic(topology=topology, population=10, generations=10, keep_portion=0.4, 
             input_dim=input_dim, output_dim=output_dim, metrics=metric)

k.runGenerations(cv=5, validation_split=0.1, n_mutations=2, crossover=0.5, 
                 performance_cap=0.1, verbose=0)

print('Total elapsed time (s): %.2f' % (time.time() - start))
```
