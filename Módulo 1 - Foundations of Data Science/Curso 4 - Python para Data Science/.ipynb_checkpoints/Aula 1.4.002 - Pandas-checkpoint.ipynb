```python
import pandas as pd


data = pd.read_csv("https://raw.githubusercontent.com/advinstai/python-datascience/master/csv/Auto.csv")

print("\n Shape: \n", data.shape)
print("\n Columns: \n", data.columns)

print("\n Info: \n", data.info)
print("\n Head: \n", data.head())
print("\n Tail: \n", data.tail(1))
print("\n Describe: \n", data.describe())

print("\n Data: \n", data)

subset = data[["mpg","cylinders"]].copy()
print("\n MPG x Cylinders: \n", subset)
subset.plot()

print("\n Índices: \n", subset.index)
```

```python
subset.set_index(pd.RangeIndex(start=0, stop=397*2, step=2), inplace=True)
print("\n Índices: \n", subset.index)

print("\n Head com novos índices: \n", subset.head(5))
display(subset[subset["cylinders"]>6].shape)
```

```python
import numpy as np
df = pd.DataFrame(data=np.flip(np.arange(10)), columns=["Seq"], index=np.arange(10))
display(df)
```

```python
df = pd.DataFrame(data=np.flip(np.arange(10)), columns=["Seq"], index=np.arange(10))
df["Seq"] = df["Seq"].astype(str).astype(int)

df["Add1"] = df["Seq"] + 1
df["Add2"] = df["Seq"].add(1)

df["Sub1"] = df["Seq"] - 1
df["Sub2"] = df["Seq"].sub(1)
df["Sub3"] = df["Seq"].sub(1)

df["Div1"] = df["Seq"] / 2
df["Div2"] = df["Seq"].div(2)
df["Div3"] = df["Seq"].divide(2)
df["Div4"] = df["Seq"].truediv(2)

df["Floor1"] = df["Seq"] // 2
df["Floor2"] = df["Seq"].floordiv(2)

df["Mod1"] = df["Seq"] % 2
df["Mod2"] = df["Seq"].mod(2)

df["Pow1"] = df["Seq"] ** 2
df["Pow2"] = df["Seq"].pow(2)

display(df)
```

```python
vals1 = pd.DataFrame(np.array([1, None, 3, 4]))
print("Impacto do None ao transformar np em pd", vals1.sum())

vals2 = pd.DataFrame(np.array([1, np.nan, 3, 4]))
print("Impacto do None ao transformar np em pd", vals2.sum())

vals3 = np.array([1, np.nan, 3, 4])
print("Impacto do NaN em um array de números", vals3.sum())
```

```python
index = [(0,0),(0,1),(1,0),(1,1)]
valor = [0, 1, 1, 2]
grid = pd.Series(valor, index=index)
display(grid)

display(grid[[i for i in grid.index if i[1] == 1]])

index = pd.MultiIndex.from_tuples(index, names=['first', 'second'])
grid = grid.reindex(index)
print(grid)
print(grid[:, 1])

grid = pd.concat([grid, grid])
display(grid)
```
