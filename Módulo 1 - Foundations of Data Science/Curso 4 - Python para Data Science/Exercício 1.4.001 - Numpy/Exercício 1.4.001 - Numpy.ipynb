```python
import numpy as np
np.set_printoptions(precision=4, suppress=True)

print("\nVetor de nulos\n", np.zeros(10))

print("\nTamanho da memória de uma matriz\n", np.zeros(10).nbytes)

v = np.zeros(10)
v[5] = 1
print("\nVetor de nulos com posição i alterada\n", v)

print("\nVetor de valores entre 10 e 49\n", np.arange(10, 50, 1))

print("\nVetor intertido\n", np.flip(np.arange(10, 50, 1)))

print("\nMatriz sequencial de 0 a 8\n", np.arange(0, 9, 1).reshape((3,3)))

print("\nMatriz identidade\n", np.identity(3))

print("\nMatriz 3x3 de valores aleatórios\n", np.random.random((3,3,3)))

m = np.random.randint(0, 100, (10,10))
print("\nMatriz 10x10\n", m)
print("\nMínimo e máximo da matriz 10x10: %.2f : %.2f\n" % (m.min(), m.max()))

v = np.random.random(30)
print("\nVetor de 30 elementos\n", v)
print("\nMédia do vetor\n", v.mean())

m = np.zeros((5,5))
for i in range(m.shape[0]):
    for j in range(m.shape[1]):
        if (i==0 or i==m.shape[0]-1) or (j==0 or j==m.shape[1]-1): m[i,j] = 1
print("\nMatriz 2D com borda\n", m)

m = np.zeros((5,5))
m[0], m[m.shape[0]-1], m[:,0], m[:,m.shape[1]-1] = 1, 1, 1, 1
print("\nMatriz 2D com borda inteligente\n", m)

n = 10
m = np.concatenate((np.zeros((1,n)),np.arange(1,n+1)*np.identity(n)), axis=0)
m = np.delete(m, m.shape[1], axis=0)
print("\nMatriz doidona\n", m)

def modify(m, r, c, ch):
    for i in range(r):
        for j in range(c):
            m[i,j,ch] = 255 if i % (j+1) < 5 and i % (j+1) > -5 else m[i,j,ch]
            m[i,j,1] = m[i,j,1] * 255
            m[i,j,1] = 255 if i == j else m[i,j,1]
            m[i,j,2] = j * m[i,j,2] * 255
    return m
f = np.full((800, 1200, 3), 1)
f = modify(f, f.shape[0], f.shape[1], 0)   
print("\nFoto\n", f)
from PIL import Image
img = Image.fromarray(f.astype('uint8'), 'RGB')
img.show()

m = np.random.random((3, 5))
print("\nMatriz antes da subtração\n", m)
print("\nMédia de cada linha\n", np.mean(m, axis=1))
m = m - np.mean(m, axis=1).reshape((m.shape[0], 1))
print("\nMatriz após a subtração\n", m)

from scipy import stats
m = np.random.randint(0, 100, (3,10))
print("\nMatriz para analisar moda\n", m)
print("\nModa desta matriz\n", np.argmax(np.bincount(m.flatten())))

print("\nMatriz a partir do arquivo\n", np.loadtxt('dados.txt'))

m = np.random.random((100,30))
np.savetxt('dados.txt', m)
print("\nMatriz salva para o arquivo\n")
```
